```python
import pandas as pd

# Load GTF file (downloaded from GENCODE)
gtf_file = "/content/drive/My Drive/3.Fall2024/Biophysics/FinalProject/dataset/gencode.v47.annotation.gtf"

# Extract gene-level data
gtf_data = pd.read_csv(gtf_file, sep="\t", comment="#", header=None)
genes = gtf_data[gtf_data[2] == "gene"]  # Filter for gene entries

# Extract relevant columns
genes_bed = genes[[0, 3, 4, 8]]  # chr, start, end, attributes
genes_bed.columns = ["chr", "start", "end", "attributes"]

# Extract gene_name from attributes
genes_bed["gene_name"] = genes_bed["attributes"].str.extract('gene_name "([^"]+)"')

# Save to BED file
genes_bed[["chr", "start", "end", "gene_name"]].to_csv("genes.bed", sep="\t", index=False, header=False)
```
